# 01 – Datenbereinigung & Explorative Datenanalyse (EDA)
Dieses Notebook enthält die Datenbereinigung und die vollständige EDA aus Aufgabe 3a.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## 1. Daten laden

```python
house_df = pd.read_csv("../data/housedata.csv")
house_df.head()
```

## 2. Erste Übersicht

```python
house_df.shape, house_df.keys()
```

## 3. Datenbereinigung

```python
# 3.1 Datentypen prüfen
house_df.info()
```

```python
# 3.2 Fehlende Werte
house_df.isna().sum()
```

```python
# 3.3 Duplikate
house_df.duplicated().sum()
house_df = house_df.drop_duplicates()
```

### 3.4 Ausreißeranalyse

```python
numeric_cols = house_df.select_dtypes(include=np.number).columns

n_cols = 3
n_rows = int(np.ceil(len(numeric_cols) / n_cols))

plt.figure(figsize=(n_cols * 4, n_rows * 3))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.boxplot(house_df[col].dropna())
    plt.title(col, fontsize=9)
    plt.xticks([])

plt.tight_layout()
plt.show()
```

## 4. Univariate Analyse

```python
num_cols = house_df.columns.drop("AboveMedianPrice")
house_df[num_cols].describe().round(3)
```

```python
house_df[num_cols].hist(figsize=(12,14))
plt.suptitle("Histogramme der numerischen Variablen")
plt.tight_layout()
plt.show()
```

### Verteilung der Zielvariablen

```python
plt.figure(figsize=(5,4))
sns.countplot(data=house_df, x="AboveMedianPrice", hue="AboveMedianPrice", palette="viridis", legend=False)
plt.title("Verteilung der Zielvariable")
plt.show()
```

## 5. Bivariate Analyse

```python
corr_target = house_df.corr()["AboveMedianPrice"]
sorted_cols = corr_target.drop("AboveMedianPrice").sort_values(ascending=False).index

plt.figure(figsize=(16, 20))

for i, col in enumerate(sorted_cols, 1):
    plt.subplot((len(num_cols) + 2) // 3, 3, i)
    sns.boxplot(data=house_df, x="AboveMedianPrice", y=col, hue="AboveMedianPrice", palette="viridis", legend=False)
    plt.title(f"{col} nach AboveMedianPrice\nKorrelation: {corr_target[col]:.2f}")
    plt.xlabel("AboveMedianPrice (0 = unter Median, 1 = über Median)")
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

## 6. Multivariate Analyse (Heatmap)

```python
corr = house_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korrelationsmatrix der Merkmale")
plt.show()
```
